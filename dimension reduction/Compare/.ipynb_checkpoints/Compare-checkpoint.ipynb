```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.decomposition import KernelPCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
import umap
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow.keras.backend as K
```

```python
df_wine = pd.read_csv('https://archive.ics.uci.edu/ml/'
                      'machine-learning-databases/wine/wine.data',
                      header=None)
```

```python
X = df_wine.iloc[:,1:]
y = df_wine.iloc[:,0]
```

```python
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
X_std
```

```python
reducer = PCA(n_components=2)
X_pca = reducer.fit_transform(X_std)

fig, ax = plt.subplots(figsize = (10,10))
ax.scatter(X_pca[:,0], X_pca[:,1], c = y, cmap = plt.cm.Set3)
plt.suptitle('PCA result', fontsize = 15)
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize = (10,10))

for idx, c in enumerate([.01, .1, 1, 10]):
    reducer = KernelPCA(n_components=2,kernel='rbf',gamma=c)
    X_kpca = reducer.fit_transform(X_std)
    ax[idx//2,idx%2].scatter(X_kpca[:,0], X_kpca[:,1], c = y, cmap = plt.cm.Set3)
    ax[idx//2,idx%2].set_title(f'c : {c}')
plt.suptitle('KernelPCA result', fontsize = 15)
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize = (10,10))

for idx, p in enumerate([1, 10, 30, 50]):
    reducer = TSNE(n_components=2,perplexity=p)
    X_tsne = reducer.fit_transform(X_std)
    ax[idx//2,idx%2].scatter(X_tsne[:,0], X_tsne[:,1], c = y, cmap = plt.cm.Set3)
    ax[idx//2,idx%2].set_title(f'perplexity : {p}')
plt.suptitle('TSNE result', fontsize = 15)
plt.tight_layout()
plt.show()
```

```python
zz = [[idx1,idx2,m_d,n_n] for idx1, m_d in enumerate([.1, .4, .7, 1.]) for idx2, n_n in enumerate([10, 30, 50, 100])]
fig, ax = plt.subplots(4,4, figsize = (20,17))


for z in zz:
    reducer = umap.UMAP(n_neighbors=z[3], n_components=2, min_dist = z[2])
    X_umap = reducer.fit_transform(X_std)
    ax[z[0],z[1]].scatter(X_umap[:,0], X_umap[:,1], c = y, cmap = plt.cm.Set3)
    ax[z[0],z[1]].set_title(f'min_dist = {z[2]} \n n_neighbor = {z[3]}')
plt.suptitle('UMAP result', fontsize = 25)
plt.tight_layout()
plt.show()
```

```python
encoder = tf.keras.models.Sequential([tf.keras.layers.Dense(2, input_shape = [X_std.shape[1]])])
decoder = tf.keras.models.Sequential([tf.keras.layers.Dense(X_std.shape[1], input_shape = [2])])
autoencoder = tf.keras.models.Sequential([encoder, decoder])

autoencoder.compile(loss = 'mse', optimizer = tf.keras.optimizers.SGD(lr = .1))
history = autoencoder.fit(X_std, X_std, epochs = 30, verbose = 0)
X_ae = encoder.predict(X_std)

fig, ax = plt.subplots(figsize = (10,10))
ax.scatter(X_ae[:,0], X_ae[:,1], c = y, cmap = plt.cm.Set3)
plt.suptitle('AutoEncoder result', fontsize = 15)
plt.tight_layout()
plt.show()
#PCA  
```

```python
stacked_encoder = tf.keras.models.Sequential([
    tf.keras.layers.Dense(7, activation = 'selu', input_shape = [X_std.shape[1]]),
    tf.keras.layers.Dense(2, activation = 'selu')
])
stacked_decoder = tf.keras.models.Sequential([
    tf.keras.layers.Dense(2, activation = 'selu', input_shape = [2]),
    tf.keras.layers.Dense(X_std.shape[1], activation = 'linear')
])
stacked_ae =  tf.keras.models.Sequential([stacked_encoder, stacked_decoder])
stacked_ae.compile(loss = 'mse', optimizer = tf.keras.optimizers.SGD(lr = .1))
history = stacked_ae.fit(X_std, X_std, epochs = 40, verbose = 0)
X_sae = stacked_encoder.predict(X_std)

fig, ax = plt.subplots(figsize = (10,10))
ax.scatter(X_sae[:,0], X_sae[:,1], c = y, cmap = plt.cm.Set3)
plt.suptitle('Stacked AutoEncoder result', fontsize = 15)
plt.tight_layout()
plt.show()
```

```python
tf.random.set_seed(42)
class Sampling(tf.keras.layers.Layer):
    def call(self, inputs):
        mean, log_var = inputs
        return K.random_normal(tf.shape(log_var),seed =42) * K.exp(log_var / 2) + mean
    
inputs = tf.keras.layers.Input(shape = [X_std.shape[1]])
z = tf.keras.layers.Dense(7, activation = 'selu')(inputs)
codings_mean = tf.keras.layers.Dense(2)(z)
codings_log_var = tf.keras.layers.Dense(2)(z)
codings = Sampling()([codings_mean, codings_log_var])
variational_encoder = tf.keras.Model(inputs = [inputs], outputs = [codings_mean, codings_log_var, codings])

decoder_inputs = tf.keras.layers.Input(shape = [2])
x = tf.keras.layers.Dense(7, activation = 'selu')(decoder_inputs)
outputs = tf.keras.layers.Dense(X_std.shape[1], activation = 'linear')(x)
variational_decoder = tf.keras.Model(inputs = [decoder_inputs], outputs = [outputs])

_, _, codings = variational_encoder(inputs)
reconstructions = variational_decoder(codings)
variational_ae = tf.keras.Model(inputs = [inputs], outputs = [reconstructions])

latent_loss = -.5 * K.sum(
    1 + codings_log_var - K.exp(codings_log_var) - K.square(codings_mean), axis = 1
)
variational_ae.add_loss(K.mean(latent_loss) / X_std.shape[1])
variational_ae.compile(loss = 'mse', optimizer = tf.keras.optimizers.SGD(lr = .1))

variational_ae.fit(X_std, X_std, epochs = 40)
```

```python
X_vae = variational_encoder.predict(X_std)[2]
X_vae_created = variational_decoder(tf.random.normal(shape = [20, 2])).numpy()
X_vae_created = variational_encoder.predict(X_vae_created)[2]

fig, ax = plt.subplots(figsize = (10,10))
ax.scatter(X_vae[:,0], X_vae[:,1], c = y, cmap = plt.cm.Set3)
ax.scatter(X_vae_created[:,0], X_vae_created[:,1], c = 'black')
plt.suptitle('Variational AutoEncoder result', fontsize = 15)
plt.tight_layout()
plt.show()
```
